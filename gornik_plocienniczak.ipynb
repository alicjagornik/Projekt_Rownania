## OPIS ZJAWISKA / PODSTAWY TEORETYCZNE

##### Zmiana masy ciała jest wynikiem różnicy między spożywanymi, a zużywanymi kaloriami. Nadmierna ilość kalorii w spożywanej diecie prowadzi do wzrostu masy ciała, podczas gdy niedobór kalorii skłania organizm do wykorzystania zapasu energii, głównie w postaci tłuszczu, do pokrycia zapotrzebowania energetycznego, co przekłada się na utratę masy ciała.

##### Ten proces idealnie odzwierciedla pierwsza zasada termodynamiki, która stwierdza, że energia nie może być ani utracona, ani stworzona, lecz może być przekształcana z jednej formy do drugiej:
$$
\large \Delta U = Q - W
$$
##### gdzie:  
- ##### $\Delta U $- zmiana energii wewnętrznej systemu,
- ##### Q - ciepło dostarczone do systemu,
- ##### W - praca wykonana przez system.

##### W kontekście zmian w masie ciała, możemy ten proces powiązać z równaniem energetycznym [[2](https://enjeck.com/blog/predict-change-differential-equations/)] :
$$\large \text{Bilans kaloryczny = kalorie spożyte - kalorie spalone} $$

##### Oznaczmy przez $c$ ilość spożytych kalorii. Aby obliczyć liczbę spalonych kalorii skorzystamy z podstawowej przemiany materii - BMR (ang. Basal Metabolic Rate). Jest to ilość energii, którą organizm zużywa w spoczynku na podstawowe funkcje życiowe, takie jak oddychanie czy regulacja temperatury ciała. Istnieje kilka metod obliczania BMR, a my wykorzystamy równanie Mifflin St Jeor, uznane za jedno z najdokładniejszych do tego celu:

$$
\large P = \left(\frac{10.0 \cdot m}{1 \text{ kg}} + \frac{6.25 \cdot h}{1 \text{ cm}} - \frac{5.0 \cdot a}{1 \text{ year}} + s \right) \frac{\text{kcal}}{\text{day}}
$$

##### gdzie:  
- ##### m - masa [kg],
- ##### h - wzrost [cm],
- ##### a - wiek [lat], 
- ##### s = +5 dla mężczyzny i s = -161 dla kobiety.

##### Jednakże BMR nie jest jedynym czynnikiem wpływającym na dzienną liczbę spalanych przez nas kalorii. Musimy uwzględnić każdą czynność zwiększającą poziom naszej aktywności, jak chociażby chodzenie pieszo na uczelnię czy ćwiczenia w domu lub na siłowni. W języku specjalistycznym nazywa się to współczynnikiem aktywności fizycznej - PAL (ang. Physical Activity Level), czyli wskaźnikiem określającym poziom aktywności człowieka.

$$
\large PAL = \frac{\text{całkowita przemiana materii}}{\text{podstawowa przemiana materii}} = f
$$

##### Obliczanie PAL wymaga jednak dość wnikliwej analizy poziomu indywidualnej, codziennej aktywności fizycznej, co dla wielu osób może być problematyczne. Dlatego, aby uczynić nasz model bardziej praktycznym i wykluczyć ewentualne błędy ludzkie, skorzystamy z uogólnych wartości, które odpowiadają podstawowym zakresom aktywności.

##### Znając zarówno nasze BMR, jak i PAL możemy "uzupełnić" nasz model. Wiemy już, że: 
 
\begin{align*}
& \large \text{kalorie spalone} = f \cdot P \\
& \large \text{kalorie zużyte} = c
\end{align*}
##### Załóżmy, że $1 \space kg$ tkanki tłuszczowej to około $7700 \space kcal$. Dlatego jeśli chcemy przybrać lub zrzucić $1 \space kg$ musimy stworzyć bilans kalorycznyny na poziomie $7700 \space kcal$. Optymalnym i zdrowym podejściem jest stopniowe osiągnięcie tego bilansu w określonym czasie. Biorąc to pod uwagę, możemy stwierdzić:

\begin{align*}
\large \Delta m(t) &= \dfrac{\text{bilans kaloryczny}}{7700} = \dfrac{c-f \cdot P}{7700} \\
\\
\large \dfrac{dm(t)}{dt} &= \dfrac{c - f \cdot \left( 10 \cdot m(t) + 6.25 \cdot h - 5 \cdot a + s \right)}{7700} \\
\\
\large m' &= \dfrac{-f \cdot m}{7700} + \dfrac{c - f \left(6.25 \cdot h - 5 \cdot a + s \right)}{7700}
\end{align*}

##### Przejdźmy teraz do rozwiązania powyższego równania różniczkowego. Chwilowo dla ułatwienia zapisu potraktujmy niewiadome $f,h,a,s$ jako stałe, gdyż nie zależą one od czasu.
##### Niech $k = \dfrac{c - f \left(6.25 \cdot h - 5 \cdot a + s \right)}{7700}$.
##### Wówczas :
$$
\large m' = \dfrac{-f \cdot m}{7700} + k
$$

##### Do rozwiązania równania wykorzystamy czynnik całkujący. Niech $ \large \mathrm{\mu} = e^{\int \frac{f}{770} \, dt} = e^{\frac{ft}{770}}$. $\space$ Wprowadzając ten czynnik do równania otrzymujemy:

\begin{align*}
\large m' + \dfrac{f}{770}m &= \large k \quad \text{/} \cdot\mathrm{\mu} \\
\\
\large \mathrm{e}^{\frac{f}{770}t} \cdot m' + \large \mathrm{e}^{\frac{f}{770}t} \cdot \small \frac{f}{770}m &= \large \mathrm{e}^{\frac{f}{770}t} \cdot k \\
\\
\left( \large \mathrm{e}^{\frac{f}{770}t} \cdot m \right)' &= \large \mathrm{e}^{\frac{f}{770}t} \cdot k
\end{align*}

##### Dalej, całkując obustronnie otrzymujemy:

\begin{align*}
\int \left( \large \mathrm{e}^{\frac{f}{770}t} \cdot m \right)' \, dt &= k \int \large \mathrm{e}^{\frac{f}{770}t} \cdot k \, dt \\
\\
\left( \large \mathrm{e}^{\frac{f}{770}t} \cdot m \right) &= \frac{770}{f} \cdot k \cdot \large \mathrm{e}^{\frac{f}{770}t} + C\\
\\
\end{align*}

##### Po przekształceniach otrzymujemy rozwiązanie w postaci:
$$ \large m(t) = \frac{770}{f} \cdot k + C \cdot \large \mathrm{e}^{- \frac{f}{770}t}$$

##### Wyznaczmy teraz wartość stałej $C$. Weźmy warunek początkowy $m(0) = m_0$, gdzie $m_0$ - masa początkowa, ponieważ na początku pomiarów nasza masa nie ulega zmianie.
$$ \large m(0) = \frac{770}{f} \cdot k + C \cdot 1 \equiv m_0$$ 
##### Zatem: $C = m_0 - \dfrac{770k}{f}$
##### Podstawiając do równania otrzymujemy:
$$ \large m(t) = \frac{770}{f} \cdot k + \left( m_0 - \dfrac{770k}{f} \right) \cdot \large \mathrm{e}^{- \frac{f}{770}t}$$
##### Pamiętajmy, że skorzystaliśmy z podstawienia $k = \dfrac{c - f \left(6.25 \cdot h - 5 \cdot a + s \right)}{7700}$.
##### Zatem wracając do oryginalnych oznaczeń otrzymujemy otatecznie równanie:
$$ \large m(t) = \frac{c - f \left(6.25 \cdot h - 5 \cdot a + s \right)}{10f} + \left( m_0 - \dfrac{c - f \left(6.25 \cdot h - 5 \cdot a + s \right)}{10f} \right) \cdot \large \mathrm{e}^{- \frac{f}{770}t}$$

## PROGRAM DO OBLICZANIA PRZEWIDYWANEJ MASY CIAŁA
##### Poniższy kod został opracowany przez autorów projektu celem obliczenia przewidywanej masy ciała oraz prognozowania zmian masy ciała na podstawie dokładnego rozwiązania równania różniczkowego. Implementacja tego rozwiązania została dokonana w języku Python, wykorzystując biblioteki takie jak:
- ##### matplotlib (generowanie wykresów,)
- ##### numpy (manipulowanie danymi numerycznymi)
- ##### math (korzystanie z funkcji matematycznych)

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

##### Głównym elementem programu jest klasa BodyWeight, której obiektami są osoby zdefiniowane przez zestaw parametrów. Wewnątrz klasy znajdują się funkcje realizujące rozwiązanie równania oraz prezentujące wyniki, takie jak:
- ##### BMR - oblicza podstawową przemianę materii (BMR) na podstawie ustalonego wzoru,
- ##### dm_dt - zmiana zmiennej m obiektu,
- ##### final_weight_analytical - prezentuje analityczne rozwiązanie równania,
- ##### all_weights - generuje wykres przedstawiający zależność masy ciała od liczby dni na podstawie rozwiązania analitycznego.
##### Funkcje final_weight_euler oraz final_weight_runge_kutta to implementacje metod numerycznych, odpowiednio: metoda Eulera oraz metoda Runge-Kutta 4 rzędu. 

```python
class BodyWeight():
    """Model do obliczania zmiany masy ciała w czasie.
    Args:
        name (str): Imię osoby.
        t (int): Czas trwania symulacji. [dni]
        m (float): Początkowa masa ciała [kg].
        h (int): Wzrost [cm].
        a (int): Wiek [lat].
        c (int): Dzienna liczba spożywanych kalorii [kcal].
        f (float): Poziom aktywności fizycznej.
        sex (str): Płeć - 'k' dla kobiety, 'm' dla mężczyzny.
    """
    def __init__(self, name, t, m, h, a, c, f, sex):
        """Funkcja inicjalizująca."""
        self.name = name
        self.t = t
        self.m = m
        self.h = h
        self.a = a
        self.c = c
        self.f = f
        if sex == "k":
            self.s = -161
        elif sex == "m":
            self.s = 5
        self.result_euler = []
        self.result_runge_kutta = []
        self.result_analytical = []
        self.time_euler = []
        self.time_runge_kutta = []
        self.time_analytical = []

    def BMR(self):
        """ Oblicz podstawową przemianę materii (BMR). """
        P = 10 * self.m + 6.25 * self.h - 5 * self.a + self.s
        return P

    def dm_dt(self, m):
        """Zmiana masy"""
        dm = (self.c - self.f * (10 * m + 6.25 * self.h - 5 * self.a + self.s)) / 7700
        return dm
    
    def final_weight_euler(self, h = 0.001):
        """ Oblicz zmianę masy ciała po upływie czasu t - metoda Eulera. """
        n = int(self.t / h)  # liczba kroków
        t = 0
        m = self.m
        for i in range(n):
            m = m + self.dm_dt(m) * h
            t = t + h
            self.time_euler.append(t)
            self.result_euler.append(m)
        return m
    
    def final_weight_runge_kutta(self, h = 0.001):
        """ Oblicz zmianę masy ciała po upływie czasu t - metoda Runge-Kutta. """
        n = int(self.t / h)  # liczba kroków
        t = 0
        m = self.m
        for i in range(n):
            k1 = h * self.dm_dt(m)
            k2 = h * self.dm_dt(m + 0.5 * k1)
            k3 = h * self.dm_dt(m + 0.5 * k2)
            k4 = h * self.dm_dt(m + k3)
            m = m + (1.0 / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)
            t = t + h
            self.time_runge_kutta.append(t)
            self.result_runge_kutta.append(m)
        return m
    
    def final_weight_analytical(self):
        """Oblicz zmianę masy ciała po upływie czasu t - implementacja metody analitycznej."""
        for t in range(1, self.t + 1):
            m = (self.m - (self.c - self.f * (6.25 * self.h - 5 * self.a + self.s)) / (10 * self.f)) * math.exp(-self.f * t / 770) + (self.c - self.f * (6.25 * self.h - 5 * self.a + self.s)) / (10 * self.f)
            self.result_analytical.append(m)
            self.time_analytical.append(t)
        return m

    def all_weights(self):
        """Wywołanie wykresu masy od czasu na podstawie rozwiązania analitycznego."""
        plt.figure(figsize=(8, 6)) 
        plt.plot(self.time_analytical, self.result_analytical) 
        plt.xlabel('Czas [dni]')
        plt.ylabel('Masa ciała [kg]')
        plt.title(f'Zmiana masy ciała - {self.name}')
        plt.grid()
        plt.show()
        
    def error(self):
        """Funkcja obliczająca błąd między rozwiązaniami analitycznym i numerycznymi."""
        r_k_e = []
        euler_e = []
        steps_e = [0.1, 0.5, 1]
        for h in steps_e:
            euler_e.append(np.abs(self.final_weight_analytical() - self.final_weight_euler(h) ))
            r_k_e.append(np.abs(self.final_weight_analytical() - self.final_weight_runge_kutta(h) ))
        
        plt.figure(figsize=(8, 6))
        plt.loglog(steps_e, euler_e, marker='o', label='Metoda:Euler')
        plt.loglog(steps_e, r_k_e, marker='o', label='Metoda: Runge-Kutta')
        plt.xlabel('Krok czasowy')
        plt.ylabel('Błąd')
        plt.title(f'Błąd między rozwiązaniami numerycznymi a analitycznym - {self.name}')
        plt.legend()
        plt.grid()
        plt.show()
```

## DOKŁADNOŚĆ ROZWIĄZAŃ NUMERYCZNYCH

##### Przykład modelu zmiany masy ciała pozwala na porównanie dokładności dwóch metod numerycznych: metody Eulera i metody Runge-Kutta.
##### Oto jak możemy porównać proponowane metody numeryczne: obliczamy maksymalną różnicę między rozwiązaniem numerycznym a analitycznym dla różnych kroków czasowych, a następnie analizujemy wartość błędu i jego rząd, czyli tangens kąta nachylenia wykresu funkcji w skali loglog do osi X.

```python
Magda.error()
Lukasz.error()
Magda_1.error()
```

##### Z wykresów wynika, że im większy krok czasowy, tym większy błąd obu metod. Kąty, jakie tworzą wykresy w stwosunku do osi $X$ są kątami ostrymi, więc im większy jest kąt tym większą wartość ma jego tangens i tym większy jest wspominany wcześniej rząd błędu. Dla kroków czasowych pomiędzy $0.1$ i $10$ tangensy kątów są porównywalne, co oznacza podobne rzędy błędu. Widać jednak jak bardzo różnią się wartość błędu, dla metody Eulera jest to maksymalnie $10^{-2}$, podczas gdy dla metody Runge-Kutta tylko $10^{-12}$. Analiza tych danych jednoznacznie wskazuje, że metoda Runge-Kutta wykazuje znacznie większą dokładność i precyzyjność w określaniu wyników.
